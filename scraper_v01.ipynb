```python
import numpy as np 
import pandas as pd 

from bs4 import BeautifulSoup
from urllib3 import request
import requests

from string import ascii_lowercase

from time import sleep
```

```python
def get_player_urls(url, index):

    response = requests.get(url)

    soup = BeautifulSoup(response.text, 'html.parser')

    url = soup.findAll("tr")[index].findAll('a')[0]['href']

    name = soup.findAll("tr")[index].findAll('a')[0].text

    url_out = "https://www.basketball-reference.com" + url

    out_dict = {name:url_out}

    # df = pd.DataFrame()
    # df['name'] = name
    # df['url'] = url_out

    return out_dict
```

```python
url_player_list = []

for letter in ascii_lowercase:
    print(letter)
    url_players = 'https://www.basketball-reference.com/players/'+letter+'/'

    response = requests.get(url_players)

    soup = BeautifulSoup(response.text, 'html.parser')
    #sleep(10)
    for player in range(1, len(soup.findAll('tr'))):
    
        try:

            url = get_player_urls(url_players, player)
            url_player_list.append(url)

        except:

            pass

```

```python
def get_dataframe(url, col_names = column_names):

    response = requests.get(url)

    soup = BeautifulSoup(response.text, 'html.parser')

    table = pd.DataFrame(soup.findAll("tr", {"class": "full_table"}),
                         columns=col_names)
    
    for col in table.columns:

        table[col] = table[col].apply(lambda x: extract_data(x))

    table['URL'] = url

    return table

            
```

```python
#Hall of fame scraper

url = "https://www.basketball-reference.com/awards/hof.html"

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
def get_hof_year(soup_tr, idx):

    hof_year = soup_tr.find('table').findAll('tr')

    year = hof_year[idx].find('th').text

    return year
```

```python
def get_hof_name(soup_tr, idx):

    hof_name = soup_tr.find('table').findAll('tr')

    name = hof_name[idx].find('td')['csk']

    return name
```

```python
def get_hof_category(soup_tr, idx):

    hof_cat = soup_tr.find('table').findAll('tr')

    category = hof_cat[idx].find('td').find('a').text

    return category
```

```python
hof_df = pd.DataFrame(columns=['Year', 'Name', 'Category'])

year = []
name = []
category = []

for i in range(2, len(soup.find('table').findAll('tr'))-1):
    
    try:
        yr = get_hof_year(soup, i)
    except:
        yr = 'na'

    try:
        nm = get_hof_name(soup, i)
    except:
        nm = 'na'

    try:
        cat = get_hof_category(soup, i)
    except:
        cat = 'na'

    year.append(yr)
    name.append(nm)
    category.append(cat)

```

```python
hof_df['Year'] = year
hof_df['Name'] = name
hof_df['Category'] = category
```

```python

hof_df = hof_df[hof_df['Year'] != '']
```
